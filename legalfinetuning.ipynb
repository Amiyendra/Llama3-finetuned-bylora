```python
from datasets import Dataset

hf_dataset = Dataset.from_list(dataset)
hf_dataset
```

```python
!pip install "unsloth[colab-new] @ git+https://github.com/unslothai/unsloth.git"

```

```python
from unsloth import FastLanguageModel

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/llama-3-8b-bnb-4bit",  # 4-bit quantized model
    max_seq_length = 2048,
    dtype = None,    # Auto dtype (bf16/float16)
    load_in_4bit = True,
)
```

```python
tokenized_dataset = hf_dataset.map(
    lambda example: tokenizer(
        f"<|system|>You are a helpful legal assistant.<|end|>\n"
        f"<|user|>{example['instruction']}<|end|>\n"
        f"<|assistant|>{example['output']}<|end|>",
        truncation=True,
        padding="max_length",
        max_length=2048,
    ),
    batched=False
)
```

```python
from transformers import DataCollatorForLanguageModeling

collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False  # For causal language modeling
)
```

```python
# Add LoRA adapters to your loaded quantized model
model = FastLanguageModel.get_peft_model(
    model,
    r = 64,
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj", "gate_proj", "up_proj", "down_proj"],
    lora_alpha = 16,
    lora_dropout = 0,
    bias = "none",
    use_gradient_checkpointing = True,
    random_state = 3407,
    use_rslora = False,
    loftq_config = None,
)
```

```python
from transformers import Trainer, TrainingArguments
from unsloth import is_bfloat16_supported

training_args = TrainingArguments(
    output_dir="./llama3-legal-finetuned",
    per_device_train_batch_size=2,
    gradient_accumulation_steps=4,
    num_train_epochs=1,
    max_steps=50,  # Train for only 50 steps to test
    logging_steps=5,
    save_steps=10,  # Save every 10 steps
    save_total_limit=2,  # Keep only last 2 checkpoints
    learning_rate=2e-4,
    fp16=True,
    bf16=is_bfloat16_supported(),
    max_grad_norm=0.3,
    warmup_ratio=0.03,
    lr_scheduler_type="constant",
    optim="paged_adamw_8bit",
    report_to="none"
)


trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    data_collator=collator,
)
```

```python
trainer.train()

```

```python
model.save_pretrained("llama3-legal-lora")
tokenizer.save_pretrained("llama3-legal-lora")
```
